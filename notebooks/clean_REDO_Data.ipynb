```python
def normalize_spotify_id(s: pd.Series) -> pd.Series:
    s = s.astype(str).str.strip()
    s = s.str.replace("spotify:track:","", regex=False)
    s = s.str.extract(r"([A-za-z0-9]{22})", expand=False).fillna(s)
    return s
```

```python
def unduplify_best(df, key):
    df = df.copy()
    df["filled_fields"] = df.notna().sum(axis=1)
    df = df.sort_values("filled_fields", ascending=False)
    df = df.drop_duplicates(subset=[key]).drop(columns=["filled_fields"])
    return df
```

```python
# Compute completion_ratio (will be NaN for songs not in catalog - that's OK)
events["completion_ratio"] = events["ms_played"] / events["duration_ms"]

# Event type: default to "play", then override
events["event_type"] = "play"
events.loc[events["skipped"] == True, "event_type"] = "skip"
# Only mark as "complete" if we have completion_ratio AND it's >= 0.8
events.loc[(events["completion_ratio"] >= 0.8) & (events["skipped"] == False), "event_type"] = "complete"

# Like proxy: handle NaN gracefully
# 1 = completed (>=80%), -1 = skipped, 0 = neutral/unknown
events["like_proxy"] = 0  # default
events.loc[events["completion_ratio"] >= 0.8, "like_proxy"] = 1
events.loc[events["skipped"] == True, "like_proxy"] = -1

# Inferred mood from song (will be NaN for unmatched songs)
events["inferred_mood"] = events["mood_label"]

print(f"Events with completion_ratio: {events['completion_ratio'].notna().sum():,}")
print(f"Events missing completion_ratio: {events['completion_ratio'].isna().sum():,}")
```

```python
events = events.sort_values("timestamp")

events["session_break"] = events["timestamp"].diff().dt.total_seconds().div(60) > 30
events["session_id"] = events["session_break"].cumsum()
```

```python
# Rolling mode function that handles NaN values
def rolling_mode(series, window=10):
    result = []
    for i in range(len(series)):
        window_vals = series.iloc[max(0, i - window + 1): i + 1].dropna()
        if len(window_vals) == 0:
            result.append(None)  # No valid values in window
        else:
            mode_result = window_vals.mode()
            result.append(mode_result.iloc[0] if len(mode_result) > 0 else None)
    return result

# Single user for now
events["user_id"] = "joseph"

# User mood state: rolling mode of recent song moods (handles NaN)
events["user_mood_state"] = rolling_mode(events["inferred_mood"].reset_index(drop=True), window=10)

print(f"user_mood_state populated: {events['user_mood_state'].notna().sum():,}")
print(f"user_mood_state missing: {events['user_mood_state'].isna().sum():,}")
```

## session_feature

```python
# ============================================
# CREATE SESSION_FEATURES TABLE
# ============================================
# Required fields:
# - user_id: Who this context belongs to
# - session_id: Listening session identifier
# - inferred_session_mood: Inferred from avg audio features of songs in session
#     (NOT self-reported — derived via rules from energy/valence/tempo/acousticness)
#     Categories: Happy / Sad / Chill / Hype / Neutral
# - activity_type: Inferred from time_of_day + platform + audio features
#     Categories: Study / Workout / Relax / Commute
# - timestamp_start: When session began
# - timestamp_end: When it ended
#
# NOTE: index.ipynb uses song-level mood_bucket (from SONG_CATALOG) for
# time_context_profiles instead of this inferred mood, because this
# inference produces ~85% neutral due to NaN defaults and narrow thresholds.
# activity_type is still used by index.ipynb for time_context_profiles.

import pandas as pd
import numpy as np

# Load data
user_events = pd.read_csv("../data/processed/user_events.csv", parse_dates=["timestamp"])
song_catalog = pd.read_csv("../data/processed/SONG_CATALOG.csv")

print(f"User events shape: {user_events.shape}")
print(f"Song catalog shape: {song_catalog.shape}")
print(f"User events columns: {user_events.columns.tolist()}")
print(f"Song catalog columns: {song_catalog.columns.tolist()}")
```

```python
# Merge user_events with song_catalog to get audio features
events_enriched = user_events.merge(
    song_catalog[["track_id", "energy", "valence", "tempo", "acousticness", "danceability", "mood_label"]],
    left_on="spotify_id",
    right_on="track_id",
    how="left"
)

print(f"Enriched events shape: {events_enriched.shape}")
print(f"Matched songs: {events_enriched['track_id'].notna().sum()} / {len(events_enriched)}")
events_enriched.head()
```

```python
# ============================================
# AGGREGATE SESSION FEATURES
# ============================================

session_features = (
    events_enriched
    .groupby(["user_id", "session_id"])
    .agg(
        # Timestamps
        timestamp_start=("timestamp", "min"),
        timestamp_end=("timestamp", "max"),
        
        # Audio feature averages for mood/activity inference
        avg_energy=("energy", "mean"),
        avg_valence=("valence", "mean"),
        avg_tempo=("tempo", "mean"),
        avg_acousticness=("acousticness", "mean"),
        avg_danceability=("danceability", "mean"),
        
        # Time of day (mode = most frequent)
        primary_time_of_day=("time_of_day", lambda x: x.mode().iloc[0] if len(x.mode()) > 0 else "unknown"),
        
        # Platform
        platform=("platform", lambda x: x.mode().iloc[0] if len(x.mode()) > 0 else "unknown"),
        
        # Existing mood data from events
        dominant_mood_label=("mood_label", lambda x: x.mode().iloc[0] if len(x.dropna().mode()) > 0 else "neutral"),
        dominant_user_mood=("user_mood_state", lambda x: x.mode().iloc[0] if len(x.dropna().mode()) > 0 else "neutral"),
        
        # Session stats
        num_tracks=("spotify_id", "count"),
        total_ms_played=("ms_played", "sum"),
    )
    .reset_index()
)

print(f"Session features shape: {session_features.shape}")
session_features.head()
```

```python
# ============================================
# DERIVE INFERRED_SESSION_MOOD
# ============================================
# Categories: Happy / Sad / Chill / Hype / Neutral
# Inferred from avg audio features per session (energy, valence, tempo, acousticness)
#
# WARNING: ~85% of sessions end up as "neutral" because:
#   - Only 2,852 of 6,731 sessions matched catalog tracks (rest have NaN features)
#   - NaN defaults to 0.5, which misses all thresholds → "neutral"
# For actual mood analysis, index.ipynb uses song-level mood_bucket instead.

def derive_mood(row):
    """
    Infer mood from avg audio features of a session:
    - Happy: high valence (>0.6), medium-high energy
    - Sad: low valence (<0.4), low energy (<0.4)
    - Chill: low energy (<0.4), high acousticness (>0.5)
    - Hype: high energy (>0.7), high tempo (>120), high danceability (>0.6)
    - Neutral: everything else (including sessions with missing audio data)
    """
    energy = row.get("avg_energy", 0.5)
    valence = row.get("avg_valence", 0.5)
    tempo = row.get("avg_tempo", 100)
    acousticness = row.get("avg_acousticness", 0.5)
    danceability = row.get("avg_danceability", 0.5)
    
    # Handle NaN values (defaults push toward "neutral")
    if pd.isna(energy): energy = 0.5
    if pd.isna(valence): valence = 0.5
    if pd.isna(tempo): tempo = 100
    if pd.isna(acousticness): acousticness = 0.5
    if pd.isna(danceability): danceability = 0.5
    
    # Hype: high energy party music
    if energy > 0.7 and tempo > 120 and danceability > 0.6:
        return "hype"
    
    # Happy: positive upbeat music
    if valence > 0.6 and energy > 0.5:
        return "happy"
    
    # Sad: melancholic slow music
    if valence < 0.4 and energy < 0.4:
        return "sad"
    
    # Chill: relaxed acoustic music
    if energy < 0.4 and acousticness > 0.5:
        return "chill"
    
    return "neutral"

session_features["inferred_session_mood"] = session_features.apply(derive_mood, axis=1)
print("Inferred session mood distribution:")
print(session_features["inferred_session_mood"].value_counts())
```

```python
# ============================================
# INFER ACTIVITY_TYPE
# ============================================
# Categories: Study / Workout / Relax / Commute
# Based on: time_of_day, platform, audio features

def infer_activity(row):
    """
    Infer activity type from context signals:
    - Workout: high energy (>0.7), fast tempo (>120), high danceability
    - Study: morning/afternoon, low-medium energy, moderate acousticness
    - Commute: mobile platform, morning or evening (typical commute times)
    - Relax: evening/night, low energy, chill/acoustic music
    """
    energy = row.get("avg_energy", 0.5)
    tempo = row.get("avg_tempo", 100)
    danceability = row.get("avg_danceability", 0.5)
    acousticness = row.get("avg_acousticness", 0.5)
    time_of_day = str(row.get("primary_time_of_day", "")).lower()
    platform = str(row.get("platform", "")).lower()
    
    # Handle NaN values
    if pd.isna(energy): energy = 0.5
    if pd.isna(tempo): tempo = 100
    if pd.isna(danceability): danceability = 0.5
    if pd.isna(acousticness): acousticness = 0.5
    
    # Check if mobile platform (likely commute/on-the-go)
    is_mobile = "ios" in platform or "android" in platform or "iphone" in platform
    
    # Workout: high energy, fast tempo, danceable
    if energy > 0.7 and tempo > 120 and danceability > 0.6:
        return "workout"
    
    # Commute: mobile + morning or evening (typical commute times)
    if is_mobile and time_of_day in ["morning", "evening"]:
        return "commute"
    
    # Study: morning/afternoon + moderate energy + some focus-friendly music
    if time_of_day in ["morning", "afternoon"] and energy < 0.6 and acousticness > 0.3:
        return "study"
    
    # Relax: evening/night + low energy + chill music
    if time_of_day in ["evening", "night"] and energy < 0.5:
        return "relax"
    
    # Default based on time
    if time_of_day in ["morning", "afternoon"]:
        return "study"
    else:
        return "relax"

session_features["activity_type"] = session_features.apply(infer_activity, axis=1)
print("Activity type distribution:")
print(session_features["activity_type"].value_counts())
```

```python
# ============================================
# FINAL SESSION_FEATURES TABLE
# ============================================

# Select required columns
session_features_final = session_features[[
    "user_id",
    "session_id",
    "inferred_session_mood",
    "activity_type",
    "timestamp_start",
    "timestamp_end"
]].copy()

# Display info
print("=" * 50)
print("SESSION_FEATURES TABLE")
print("=" * 50)
print(f"\nColumns:")
for col in session_features_final.columns:
    print(f"  - {col}")
print(f"\nShape: {session_features_final.shape}")
print(f"\nData types:\n{session_features_final.dtypes}")

session_features_final.head(10)
```
