```python
from collections import defaultdict

def build_index(df, field):
    index = defaultdict(list)
    for _, row in df.iterrows():
        key = row[field]
        index[key].append(row["track_id"])
    return dict(index)

genre_index = build_index(SONG_CATALOG, "genre")
mood_index = build_index(SONG_CATALOG, "mood_bucket")
energy_index = build_index(SONG_CATALOG, "energy_label")
```

```python
# Brute force (slow)
brute = SONG_CATALOG[
    (SONG_CATALOG["mood_bucket"] == "sad") &
    (SONG_CATALOG["energy_label"] == "calm")
]["track_id"].tolist()
```

```python
indexed = list(set(mood_index["sad"]) & set(energy_index["calm"]))

print("Brute count:", len(brute))
print("Index count:", len(indexed))
print("Same results?", set(brute) == set(indexed))
```

```python
def retrieve_candidates(genres=None, mood=None, energy=None):
    sets = []

    if genres:
        genre_hits = set()
        for g in genres:
            genre_hits |= set(genre_index.get(g, []))
        sets.append(genre_hits)

    if mood:
        sets.append(set(mood_index.get(mood, [])))

    if energy:
        sets.append(set(energy_index.get(energy, [])))

    if not sets:
        return []

    return list(set.intersection(*sets))
```

```python
cands = retrieve_candidates(
    genres=["hip-hop", "pop"],
    mood="happy",
    energy="energetic"
)

len(cands)
```
