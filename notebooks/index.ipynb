## Index Fields

**Indexed fields (from SONG_CATALOG, used for Stage 1 retrieval):**
- `track_id` — primary key
- `genre` — 114 genres
- `artist_name` — tokenized
- `title` — tokenized
- `mood_bucket` — 5 categories: happy, sad, chill, hype, focus
- `energy_label` — 3 levels: calm, medium, energetic

**Non-indexed (used in Stage 2 ranking):**
- `popularity`
- `energy`, `valence`, `danceability`, `acousticness`, `tempo` (raw audio features)
- User audio preferences from `USER_PROFILE.csv` (energy_pref, valence_pref, etc.)
- Time-context defaults from `time_context_profiles.json`

**USER_PROFILE.csv schema:**
- `user_id`, `top_genres`, `energy_pref`, `valence_pref`, `danceability_pref`
- `acousticness_pref`, `tempo_pref`, `mood_bias`, `activity_preferences`, `onboarding_source`

````python
from collections import defaultdict

def build_index(df, field):
    index = defaultdict(list)
    for _, row in df.iterrows():
        key = row[field]
        index[key].append(row["track_id"])
    return dict(index)

# Category indexes (exact match)
genre_index = build_index(SONG_CATALOG, "genre")
mood_index = build_index(SONG_CATALOG, "mood_bucket")
energy_index = build_index(SONG_CATALOG, "energy_label")

# Tokenized indexes (for text search on title and artist)
def build_token_index(df, field):
    index = defaultdict(list)
    for _, row in df.iterrows():
        text = str(row[field]).lower().strip()
        tokens = text.split()
        for token in tokens:
            index[token].append(row["track_id"])
    return dict(index)

title_index = build_token_index(SONG_CATALOG, "title")
artist_index = build_token_index(SONG_CATALOG, "artist_name")

print(f"genre:  {len(genre_index)} keys")
print(f"mood:   {len(mood_index)} keys")
print(f"energy: {len(energy_index)} keys")
print(f"title:  {len(title_index)} tokens")
print(f"artist: {len(artist_index)} tokens")
````

````python
def retrieve_candidates(genres=None, mood=None, energy=None):
    sets = []

    if genres:
        genre_hits = set()
        for g in genres:
            genre_hits |= set(genre_index.get(g, []))
        sets.append(genre_hits)

    if mood:
        sets.append(set(mood_index.get(mood, [])))

    if energy:
        sets.append(set(energy_index.get(energy, [])))

    if not sets:
        return []

    return list(set.intersection(*sets))
````

## Time-Context Profiles

Build per-user, per-timeslot profiles from historical events.
This lets the backend auto-suggest mood/activity based on time of day.

Uses:
- **USER_EVENTS**: which track was played, when, skipped or not, time_of_day
- **SESSION_FEATURE**: activity_type per session
- **SONG_CATALOG**: track_id -> genre and mood_bucket mapping

Mood is derived from **song-level `mood_bucket`** (what music the user actually listened to),
NOT from the session-level `self_reported_mood` (which was 85% neutral due to inference rules).

Output: `indexes/time_context_profiles.json` with structure:
```json
{
  "user_id": {
    "morning": {
      "typical_mood": "focus",
      "mood_distribution": {"focus": 0.49, "sad": 0.23, ...},
      "typical_activity": "commute",
      "activity_distribution": {"commute": 0.82, "workout": 0.10, ...},
      "top_genres": ["indie", "pop", ...],
      "event_count": 135,
      "confidence": 0.053
    },
    "afternoon": { ... },
    "evening": { ... },
    "night": { ... }
  }
}
```

````python
import json

events = USER_EVENTS.copy()

# Merge events with sessions to get activity_type per event
merged = events.merge(
    SESSION_FEATURE[["user_id", "session_id", "activity_type"]],
    on=["user_id", "session_id"],
    how="left",
)

# Map track_id -> genre and mood_bucket from catalog (song-level, not session-level)
genre_map = SONG_CATALOG.set_index("track_id")["genre"].to_dict()
mood_map = SONG_CATALOG.set_index("track_id")["mood_bucket"].to_dict()
merged["genre"] = merged["track_id"].map(genre_map)
merged["song_mood"] = merged["track_id"].map(mood_map)

# Filter to meaningful events (played, not skipped)
meaningful = merged[
    (merged["event_type"] == "play") &
    (~merged["skipped"])
]

# Build time-context profiles
time_context_profiles = {}

for user_id, user_events in meaningful.groupby("user_id"):
    user_ctx = {}
    for tod, tod_events in user_events.groupby("time_of_day"):
        # Mood from the SONGS actually listened to (catalog mood_bucket)
        mood_dist = tod_events["song_mood"].dropna().value_counts(normalize=True)
        typical_mood = mood_dist.index[0] if len(mood_dist) > 0 else "focus"

        # Activity from sessions
        activities = tod_events["activity_type"].dropna()
        activity_dist = activities.value_counts(normalize=True) if len(activities) > 0 else pd.Series()
        typical_activity = activity_dist.index[0] if len(activity_dist) > 0 else None

        # Top genres by play count
        genres = tod_events["genre"].dropna()
        top_genres = genres.value_counts().head(5).index.tolist() if len(genres) > 0 else []

        confidence = len(tod_events) / len(user_events)

        user_ctx[tod] = {
            "typical_mood": typical_mood,
            "mood_distribution": {m: round(float(v), 3) for m, v in mood_dist.items()},
            "typical_activity": typical_activity,
            "activity_distribution": {a: round(float(v), 3) for a, v in activity_dist.items()} if len(activity_dist) > 0 else {},
            "top_genres": top_genres,
            "event_count": int(len(tod_events)),
            "confidence": round(confidence, 3),
        }

    uid = user_id.lower() if isinstance(user_id, str) else str(user_id)
    time_context_profiles[uid] = user_ctx

print(f"Built time context profiles for {len(time_context_profiles)} users\n")
for uid, ctx in time_context_profiles.items():
    print(f"{uid}:")
    for tod, info in sorted(ctx.items()):
        print(f"  {tod:10s}  mood={info['typical_mood']:8s}  activity={str(info['typical_activity']):8s}  "
              f"genres={info['top_genres'][:3]}  confidence={info['confidence']}")
        print(f"             mood_dist={info['mood_distribution']}")
        print(f"             activity_dist={info['activity_distribution']}")
````

````python
# Save time context profiles to indexes/
with open(indexes_dir / "time_context_profiles.json", "w") as f:
    json.dump(time_context_profiles, f, indent=2)

print(f"Time context profiles saved to {(indexes_dir / 'time_context_profiles.json').resolve()}")
````
